```python
from codebase.viz import plot_binary_data
from codebase.dataset import ClassificationDataset
from codebase.model import KNNClassifier
from codebase.evaluation import EvaluationTools
```

```python
dataset = ClassificationDataset(n_classes=5, train_size=0.8)
```

```python
knn = KNNClassifier(k=5)
knn.fit(dataset.X_train, dataset.y_train)
y_pred = knn.predict(dataset.X_test)
```

```python
evaluation = EvaluationTools(y_pred=y_pred, y_true=dataset.y_test)
evaluation.accuracy_score()
evaluation.f1_score()
```

```python
plot_binary_data(dataset.X_test, y_pred)
```
